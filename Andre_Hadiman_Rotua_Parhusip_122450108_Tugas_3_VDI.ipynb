- Tujuan: Ingin menampilkan perbandingan tingkat literasi berdasarkan kelompok usia dari tahun 2019-2023

- User yang dituju adalah n user global

- Meskipun tone tidak secara eksplisit dinyatakan dalam kode, dari elemen-elemen visualisasi dapat disimpulkan bahwa tonenya adalah:
Informatif: Menyajikan data secara jelas dan terstruktur
Objektif: Menampilkan data tanpa bias atau interpretasi subjektif
Profesional: Menggunakan desain yang bersih dan mudah dibaca
Edukatif: Bertujuan untuk memberikan pemahaman tentang tren literasi digit

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = [2019, 2020, 2021, 2022, 2023]
age_groups = ['18-25', '26-35', '36-45', '46-55', '56+']
data = np.array([
    [75, 80, 85, 88, 90],  # 18-25
    [70, 75, 80, 84, 87],  # 26-35
    [60, 65, 72, 78, 82],  # 36-45
    [50, 55, 62, 68, 73],  # 46-55
    [30, 35, 42, 48, 55]   # 56+
])

# Membuat plot
plt.figure(figsize=(12, 8))
for i, age_group in enumerate(age_groups):
    plt.plot(years, data[i], marker='o', linewidth=2, markersize=8, label=age_group)

# Kustomisasi plot
plt.title('Tren Literasi Digital di Indonesia berdasarkan Kelompok Usia (2019-2023)', fontsize=16)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Tingkat Literasi Digital (%)', fontsize=12)
plt.legend(title='Kelompok Usia', title_fontsize='12', fontsize='10', loc='lower right')
plt.grid(True, linestyle='--', alpha=0.7)

# Mengatur batasan sumbu y dari 0 hingga 100
plt.ylim(0, 100)

# Menambahkan anotasi untuk highlight
plt.annotate('Kesenjangan\nterbesar', xy=(2023, 72.5), xytext=(2021, 60),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10)

# Menyimpan plot
plt.tight_layout()
plt.savefig('tren_literasi_digital_indonesia.png', dpi=300)
plt.show()
```
